```python
# create a plotting function

def plot_predictions(train_data=x_train, train_labels=y_train,
                    test_data=x_test, test_labels=y_test, predictions=y_pred):
    plt.figure(figsize=(10, 7))
    plt.scatter(train_data, train_labels, c="b", label="Training Data")
    plt.scatter(test_data, test_labels, c="g", label="Testing Data")
    plt.scatter(test_data, predictions, c="r", label="Predictions")
    plt.legend();
```

```python
plot_predictions(train_data=x_train,
                 train_labels=y_train,
                 test_data=x_test,
                 test_labels=y_test,
                 predictions=y_pred
)
```

```python
mae = tf.metrics.mean_absolute_error(y_test, tf.squeeze(tf.constant(y_pred)))
mae
```

```python
mse = tf.metrics.mean_squared_error(y_test, tf.squeeze(tf.constant(y_pred)))
mse
```

```python
def mae(y_true, y_pred):
    return tf.metrics.mean_absolute_error(
        y_true=y_true,
        y_pred=tf.squeeze(tf.constant(y_pred))
    )
```

```python
def mse(y_true, y_pred):
    return tf.metrics.mean_squared_error(
        y_true=y_true, 
        y_pred=tf.squeeze(tf.constant(y_pred))
    ) 
```

## A larger Example

```python
# import required library
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
insurance = pd.read_csv("https://raw.githubusercontent.com/stedy/Machine-Learning-with-R-datasets/master/insurance.csv")
insurance.info()
```

```python
insu_one_hot = pd.get_dummies(insurance)
insu_one_hot
```

```python
x = insu_one_hot.drop("charges", axis=1)
y = insu_one_hot["charges"]

y, x
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, 
                                                    random_state=42)

len(x_train), len(x_test), len(y_train), len(y_test)
```

```python
x_train = tf.constant(x_train)
y_train = tf.constant(y_train)
x_test = tf.constant(x_test)
y_test = tf.constant(y_test)

# type(x_train)
```

```python
tf.random.set_seed(42)

insu_model = tf.keras.Sequential([
    tf.keras.layers.Dense(100, activation="relu"),
    tf.keras.layers.Dense(100, activation="relu"),
    tf.keras.layers.Dense(100, activation="relu"),
    tf.keras.layers.Dense(1)
])

insu_model.compile(loss=tf.keras.losses.mae,
                  optimizer=tf.keras.optimizers.Adam(learning_rate=0.1),
                  metrics=["mae"])

history = insu_model.fit(x_train, y_train, epochs=200)
```

```python
insu_model.evaluate(x_test, y_test)
```

```python
pd.DataFrame(history.history).plot()
plt.ylabel("loss")
plt.xlabel("epochs")
```

## Preprocessing Data

```python
insu_one_hot["age"].plot(kind="hist")
```

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder

# Column transformer
ct = make_column_transformer(
    (MinMaxScaler(), ["age", "bmi", "children"]),
    (OneHotEncoder(handle_unknown="ignore"), ["sex", "smoker", "region"])
)

X = insurance.drop("charges", axis=1)
Y = insurance["charges"]

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, 
                                                    random_state=42)

ct.fit(X_train)

X_train_normal = ct.transform(X_train)
X_test_normal = ct.transform(X_test)
```

```python
X_train.loc[0]
```

```python
X_train_normal = tf.constant(X_train_normal)
Y_train = tf.constant(Y_train)
X_test_normal = tf.constant(X_train_normal)
Y_test = tf.constant(Y_train)
```

```python
X_train.shape, X_train_normal.shape
```

```python
tf.random.set_seed(42)

insu_model = tf.keras.Sequential([
    tf.keras.layers.Dense(100, activation="relu"),
    tf.keras.layers.Dense(100, activation="relu"),
    tf.keras.layers.Dense(100, activation="relu"),
    tf.keras.layers.Dense(1)
])

insu_model.compile(loss=tf.keras.losses.mae,
                  optimizer=tf.keras.optimizers.Adam(learning_rate=0.1),
                  metrics=["mae"])

history = insu_model.fit(X_train_normal, Y_train, epochs=200)
```

```python
insu_model.evaluate(X_test_normal, Y_test)
```

```python
pd.DataFrame(history.history).plot()
plt.ylabel("loss")
plt.xlabel("epochs")
```
